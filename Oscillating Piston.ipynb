```python
calibration=0.238/9 * u.mV/u.kPa/u.V

def openfile(filename):
    with open(filename) as f:
        rawdata=list(csv.reader(f))
        f.close() 
    return rawdata
def osc_funct(t, P_0, C, k, omega, phi):
    P=P_0+C*np.exp(-k*t/2)*np.cos(omega*t+phi)
    return P

def kappa(omega,k,m,V,P0,A):
    kappa=(omega**2+((k**2)/4))*((m*V)/(P0*A**2))
    return kappa

def kaper2(omega,k,P0,errorom,errork,errorP0):
    m_err=0.0005
    A_err=area_err.magnitude
    factor=(massplung.magnitude*totvol.magnitude)/(P0*pistarea.magnitude**2)
    expr1=(omega**2)+((k**2)/4)
    root=(2*omega*errorom)**2+((k/2)*errork)**2+(expr1*m_err)**2+((expr1/P0)*errorP0)**2+((2*expr1*A_err)/pistarea.magnitude)**2
    kappa_error=factor*np.sqrt(root)
    print("Kappa Error: " + str(kappa_error))


def kappaerror(omega,k,P0,m,errorom,errork,errorP0):
    #no dV--constant; see sample calculations for partial deriv explanation
    #m=massplunger.magnitude+weights
    #V=totvol.magnitude
    #A=pistarea.magnitude
    m_err=0.0005
    A_err=area_err.magnitude
    dk=(k *(massplung.magnitude+m)*totvol.magnitude)/(2*P0*pistarea.magnitude**2)
    dom=(2*omega*(massplung.magnitude+m)*totvol.magnitude)/(P0*pistarea.magnitude**2)
    dP0=-(omega**2 +(k**2/4))*((massplung.magnitude+m)*totvol.magnitude)/(P0**2*pistarea.magnitude**2)
    dA=-2*(omega**2+(k**2/4))*((massplung.magnitude+m)*totvol.magnitude)/(P0*pistarea.magnitude**3)
    dm=(omega**2+(k**2/4))*(totvol.magnitude)/(P0*pistarea.magnitude**2)
    kapp_err=np.sqrt((dom*errorom)**2+(dk*errork)**2+(dP0*errorP0)**2+(dA*A_err)**2+(dm*m_err)**2)
    print("Kappa Error: " +str(kapp_err))

def analysis(filename,x,title,mass_weight):
    #strip file
    file=openfile(filename)
    header = file[:7]
    gasdata = file[7:]
    for _ in gasdata:
        del _[1]
    for i in gasdata:
        i[1]=i[1].replace(",", ".")
    gasdata=np.array(gasdata,dtype=float)
    #analyzing data from trials
    time=(gasdata[:,0])/1000
    volt=gasdata[:,1]
    corrected_volt=(volt*u.mV)/(9*u.V)
    pressure=corrected_volt/calibration
    pressure=pressure.magnitude
    start_at=np.argmax(pressure)+x
    end_at=start_at+500
    analysisP=pressure[start_at:end_at]
    analysist=time[start_at:end_at]-time[start_at]
    plt.plot(analysist, analysisP, marker='.', linestyle="", label=f'{title}')
    p0 = [np.mean(analysisP), np.max(analysisP), -2*np.log10(0.5)/0.1, 2*np.pi/0.03, 0.5*np.pi]
    fit_param=scipy.optimize.curve_fit(osc_funct,analysist, analysisP, p0=p0)
    filler=np.linspace(analysist[0], analysist[-1], 1000)
    plt.plot(filler, osc_funct(filler, *fit_param[0]), label='Best Fit')
    P_0, C, k_val, omega_val, phi = fit_param[0]
    kappa_val=kappa(omega_val,k_val,massplung.magnitude+mass_weight,totvol.magnitude,(P_0+101.325)*1000,pistarea.magnitude)
    print("Optimized Parameters:" +"P0: " +str(P_0)+ ", C: "+str(C)+", k: "+ str(k_val)+", omega: "+str(omega_val)+", phi: "+str(phi))
    print("Kappa Value is: "+str(kappa_val))
    plt.xlabel('Time (s)')
    plt.ylabel('Pressure (kPa)')
    plt.legend()
    plt.show()
```
